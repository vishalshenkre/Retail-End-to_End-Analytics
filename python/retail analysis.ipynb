```python
df=df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
```

```python
df.columns
```

```python
df['order_date']= pd.to_datetime(df['order_date'])
```

```python
df.info()
```

```python
df['order_year']= df['order_date'].dt.year
df['order_month']= df['order_date'].dt.month
df['month_name']= df['order_date'].dt.month_name()
df['quarter']= df['order_date'].dt.quarter
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['customer_name'].fillna("unknown_customer",inplace=True)
```

```python
df['state'].fillna("unknown",inplace=True)
```

```python
df['profit'].fillna(df["profit"].median(),inplace=True)
```

```python
df["month_name"]=df["order_date"].dt.month_name()
```

```python
df.isnull().sum()
```

```python
df['sales'].fillna(df["sales"].median(),inplace=True)
```

```python
df['sales'].describe().round(2)
```

```python
Q1=df["sales"].quantile(0.25)
Q3=df["sales"].quantile(0.75)
IQR=Q3-Q1

lower=Q1-1.5*IQR
upper=Q3+1.5*IQR

len(df[(df["sales"]<lower)  |
(df["sales"]>upper)])
```

```python
#profit margin 
df["profit_margin"]=(df["profit"]/df["sales"])*100
```

```python
#sales_category
#(low/medium/high)
df["sales_category"]= pd.qcut(df["sales"],     
                     q=3,
labels=["Low","Medium","High"])
```

```python
df.head()
```

```python
#discount category
df["discount_categoty"]=np.where(df["discount"]>0.3,"High discount",
                                 "Low Discount")
```

# Group By

## Total sales by region


```python
df.groupby("region")["sales"].sum().sort_values(ascending=False).astype(float).round(2)
```

## profit by category

```python
df.groupby("category")["profit"].sum().sort_values(ascending=False).astype(float).round(2)
```

## Sales by segment

```python
df.groupby("segment")["sales"].sum().sort_values(ascending=False).astype(int)
```

## Profit margin by sales category

```python
df.groupby('sales_category')["profit_margin"].mean()
```

## discount impact on profit

```python
df.groupby("discount_categoty")["profit"].mean().round(2)
```

# pivot table


### sales by region and category


```python
pivot1=pd.pivot_table(df,
                      values="sales",
                      index="region",
                      columns="category",
                      aggfunc="sum").round(2)
pivot1
```

### profit margin pivot

```python
pivot2=pd.pivot_table(df,
                      values="profit_margin",
                      index="segment",
                      columns="sales_category",
                      aggfunc="mean").round(3)
pivot2
            
```

# numpy


```python
df["profit_status"] = np.where(df["profit"]<0,
                               "loss",
                               "profit")
```

```python
df["large_order"]=np.where(df["sales"]>df["sales"].mean(),
                           "above Average",
                           "below average")
```

```python
df.head()
```

# sorting

### top 10 customers by sales

```python
df.groupby("customer_name")["sales"]\
    .sum().round(2)\
    .sort_values(ascending=False)\
    .head(10)
```

### Top 5 loss making orders

```python
df.sort_values("profit").head()
```

### bottom 5 customers sales


```python
df.groupby("customer_name")["sales"]\
    .sum().round(2)\
    .sort_values()\
    .head(5)
```
